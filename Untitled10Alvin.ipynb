```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_weights = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_weights, config_file)
```

```python
classLabels = []
file_name = 'labels.txt'
with open(file_name, 'rt') as fpt:
  classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True)
```

```python
img = cv2.imread('pas_foto.jpg')
plt.imshow(img)
```

```python
ClassIndex, confidece, bbox = model.detect(img, confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
  cv2.rectangle(img, boxes, (255, 40, 255), 5)
  cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(2, 255, 0), thickness=10)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    cap = cv2.VideoCapture(-1)
if not cap.isOpened():
    raise IOError("Cannot open the video web camera")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    if not ret:
        break

    ClassIndex, confidece, bbox = model.detect(frame, confThreshold=0.5)

    print(ClassIndex)

    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if ClassInd <= 80:
              cv2.rectangle(frame, boxes, (0, 0, 255), 3)
              cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+30), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)

    cv2.imshow('Fix Ada BOXNYA!', frame)

    if cv2.waitKey(2) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
